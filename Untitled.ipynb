
# Part 1 Cleaning the data


## Clean Player

```python
df.Player
```

```python
df['Player'].value_counts()
```


## Clean Pos

```python
df['Pos'] = df['Pos'].str.strip()
```

```python
df['Pos'] = df['Pos'].replace('SF.','SF')
df['Pos'] = df['Pos'].replace('sg','SG')
df['Pos'] = df['Pos'].replace('PFa','PF')
df['Pos'] = df['Pos'].replace('Pg','PG')
df['Pos'] = df['Pos'].replace('SGa','SG')
df['Pos'] = df['Pos'].replace('Sf','SF')
```

```python
df.Pos
```

```python
df['Pos'].value_counts()
```


## Cleaning Age

```python
df['Age'] = df['Age'].replace(-19,19)
df['Age'] = df['Age'].replace(280,28)
```

```python
df.Age
```

```python
df['Age'].value_counts()
```


## Clean Team

```python
df['Tm'] = df['Tm'].str.strip()
```

```python
df['Tm'] = df['Tm'].replace('NyK','NYK')
df['Tm'] = df['Tm'].replace('H0U','HOU')
```

## Top 5 player with total points

```python
Top5PTS= df['3P']*3 + df['2P'] *2 +df['FT'] 
```

```python
Top5PTS = df.nlargest(5, 'PTS')
```

```python
Top5PTS[['Player','2P','3P','FT','PTS']]
```

```python
Top5PTS.loc[:,['Player','2P','3P','FT']].plot.barh(x='Player',stacked=True, figsize=(14,4))
plt.title('Total Points composistion')
plt.xlabel('Points')
```

```python
Top5PTS.plot.barh(x='Player',y='PTS',color = 'Red',figsize=(14,4))
plt.title('Total points')
plt.xlabel('Points')
```
